```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
df = pd.read_csv("ecommerce - ecommerce.csv")
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
df.head()
```

```python
df.info()
```

```python
# 1- What are the total Order value for each country
df.groupby(["Country"])["OrderValue"].sum().reset_index()
```

```python
# 2- What each Major Category sold quantity during November 2011

df[df["Year-Month"]=="2011-11"].groupby(["Major Category"])["Quantit"].sum().reset_index()
```

```python
# 3- What's Total Order Value for each Minor Item in Kitchen
df_kitchen=df[df["Major Category"]=="Kitchen"]
df_kitchen.groupby(["Minor Category"]).sum()
```

```python
df["Year-Month"].value_counts()
```

```python
# 4- What Month/s contributes to high demand in orders
df.groupby(["Year-Month"])["Quantit"].sum()
```

```python
# 5- What Month/s contributes to high OrderValue?
df.groupby(["Year-Month"])["OrderValue"].sum().reset_index()
```

```python
df.head()
```

```python
# 6- Why April and December suffer from low sales? 

df2=df[(df["Year-Month"]=="2011-04") |(df["Year-Month"]=="2011-12")]
df2.groupby(["Major Category","Minor Category"])["Quantit","OrderValue","Year-Month"].sum().reset_index()
```

```python
# 7- Why May and November have high sales? 
df3=df[(df["Year-Month"]=="2011-05") |(df["Year-Month"]=="2011-11")]
df2.groupby(["Major Category","Minor Category"])["Quantit","OrderValue","Year-Month"].sum().reset_index()
```

```python
# 8- Most Major Items sold in April and December
df[(df["Year-Month"]=="2011-04") |(df["Year-Month"]=="2011-12")].groupby(["Major Category"]).sum()
```

```python
# 9- Most Major Items sold in May and November
df[(df["Year-Month"]=="2011-05") |(df["Year-Month"]=="2011-11")].groupby(["Major Category"]).sum()
```

```python
df["hour"]=pd.to_datetime(df["InvoiceTime"]).dt.hour
```

```python
# 10- What Hour recives high demand for ordering?
df.groupby(["hour"])["Quantit"].sum().reset_index()
```

```python
def time(x):
    if x in range (7,12):
        return "Morning"
    elif x in range (12,16):
        return "Afternoon"
    else:
        return "Evening"
df["time"]=df["hour"].apply(lambda x:time(x))
```

```python
df.head()
```

```python
df.groupby(["time"])["Quantit"].sum()
```

```python
# 12 - Order VAlue For each Country in each month
df.pivot_table(values="OrderValue",index="Year-Month",columns="Country",aggfunc="sum")
```

```python
# 13 - Order VAlue For each Country and Major Category in Minor Category
df.pivot_table(values="OrderValue",index=["Country","Major Category"],columns="Year-Month",aggfunc="sum")
```

```python
# 14 - Order VAlue For each Country and Major Category in Minor Category
df.pivot_table(values="OrderValue",index=["Country","Major Category"],columns="hour",aggfunc="sum")
```

```python
# 15- Correlation between 'Quantit','OrderValue','UnitPrice'.
df[['Quantit','OrderValue','UnitPrice']].corr()
```
